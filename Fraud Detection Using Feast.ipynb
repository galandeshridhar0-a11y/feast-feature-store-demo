# Fraud Detection Using Feast

## Use Case

Imagine a customer using a credit card to make a transaction.

For example:

* **Customer ID:** 101
* **Transaction Amount:** ₹50,000
* **Transaction Location:** Mumbai

The bank's fraud detection system must quickly decide:

> **Is this transaction legitimate or fraudulent?**

To make this decision, the machine learning model needs information about the customer's recent behavior.

---

## Features Used by the Fraud Detection Model

The system calculates features such as:

| Feature                                 |    Value |
| --------------------------------------- | -------: |
| Transactions in the last 10 minutes     |        8 |
| Total amount spent in the last 24 hours | ₹200,000 |
| Distance from the previous transaction  |   500 km |

These features help the model identify unusual behavior.

For example:

* The customer made **8 transactions in only 10 minutes**.
* The customer has already spent **₹200,000 in the last 24 hours**.
* The new transaction is **500 km away from the previous transaction**.

Together, these features may indicate suspicious activity.

---

## How Feast Helps

Feast acts as a **Feature Store**.

It helps the system:

1. Receive new transaction data.
2. Calculate or update useful machine learning features.
3. Store the latest feature values.
4. Provide those features quickly to the fraud detection model.

The model does not need to directly process all raw transaction history every time.

Instead, it can request the latest features for:

```text
customer_id = 101
```

Feast can then return:

```text
transactions_last_10_minutes = 8
amount_last_24_hours = ₹200,000
distance_from_last_transaction = 500 km
```

The model uses these values to make a prediction.

---

## Model Prediction

The features are converted into a feature vector:

```text
[8, 200000, 500]
```

This feature vector is passed to the fraud detection model.

The model may return:

```text
Fraud Probability = 98%
```

The application can then:

```text
IF fraud probability is high:
    Block the transaction
ELSE:
    Approve the transaction
```

---

## Complete Business Flow

```text
New Transaction
        ↓
Feature Engineering
        ↓
Updated Customer Features
        ↓
Feast Online Store
        ↓
Retrieve Latest Features
        ↓
Fraud Detection Model
        ↓
Fraud or Legitimate
```

### Key Idea

> **Feast does not detect fraud by itself. Feast manages and serves the latest machine learning features that the fraud detection model needs to make a fast and accurate prediction.**
